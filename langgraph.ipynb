```python
def model1(input):
    return input + " hi"

def model2(input):
    return input + " there"
```

```python
from langgraph.graph import Graph

graph = Graph()

graph.add_node("node_1", model1)
graph.add_node("node_2", model2)

graph.add_edge("node_1","node_2")

graph.set_entry_point("node_1")
graph.set_finish_point("node_2")

app = graph.compile()
```

```python
app.invoke("hello")
```
